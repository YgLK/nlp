```python
import requests

target_url = 'https://www.ic.unicamp.br/~stolfi/voynich/mirror/reeds/docs/FSG.txt'

response = requests.get(target_url)
data = response.text
```

```python
print(data)
```

```python
sep = '# page 119 and 120'
valid_text = data.split(sep, 1)[0]
# remove '-' and '=' 
valid_text = valid_text.replace('-', '').replace('=', '')
print(valid_text)
```

```python
# filter out the comments and blank lines
text = [x for x in valid_text.split("\n") if not x.startswith('#') and x != '\x0c']
text
```

```python
count_dict = {}
words_in_text_count = 0

for line in text:
  for word in line.split(','):
      if word not in count_dict.keys():
        count_dict[word] = 1
      else:
        count_dict[word] += 1
      words_in_text_count += 1

# print(count_dict)
print(f"Count of words: {words_in_text_count}")
```

```python
count_dict
```

```python
sorted_words_count = dict(reversed(sorted(count_dict.items(), key=lambda item: item[1])))
sorted_words_count
```

```python
import numpy as np

x_rank = list(range(1, len(sorted_words_count)+1))
y_freq = np.array(list(sorted_words_count.values())) / words_in_text_count
y_freq
```

```python
import matplotlib.pyplot as plt

plt.plot(x_rank, y_freq)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.xscale('log') 
plt.yscale('log') 

plt.plot(x_rank, y_freq)
plt.show()
```

```python
logx, logy = np.log(x_rank), np.log(y_freq)

p = np.polyfit(logx, logy, 1)
y_fit = np.exp(np.polyval(p, logx))

# best fitting line
plt.plot(x_rank, y_fit, 'r-')
# original data
plt.plot(x_rank, y_freq)
plt.yscale('log')
plt.xscale('log')
plt.show()
```

```python
import networkx as nx
from networkx.algorithms import bipartite

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (40,20)
```

```python
words_for_graphL={}
for k in sorted_words_count_cz.keys():
  if sorted_words_count_cz[k]>40:
    words_for_graphL[k]=sorted_words_count_cz[k]
words_for_graphL = dict(reversed(sorted(words_for_graphL.items(), key=lambda item: item[1])))
print(len(words_for_graphL))
```

```python
words_for_graphR={}
for k in words_for_graphL.keys():
  words_for_graphR[k.upper()]=words_for_graphL[k]
words_for_graphR = dict(reversed(sorted(words_for_graphR.items(), key=lambda item: item[1])))
```

```python
G = nx.Graph()
G.add_nodes_from(words_for_graphL.keys(), bipartite=0)
G.add_nodes_from(words_for_graphR.keys(), bipartite=1)
```

```python
word_list= text_cz.split(",")
for i in range(len(word_list)-2):
  if words_for_graphL.get(word_list[i]) is not None and words_for_graphL.get(word_list[i+1]) is not None:
    G.add_edges_from([(word_list[i], word_list[i+1].upper())])
  if words_for_graphL.get(word_list[i]) is not None and words_for_graphL.get(word_list[i+2]) is not None:
    G.add_edges_from([(word_list[i], word_list[i+2].upper())])
```

```python
nx.draw_networkx(G, pos = nx.drawing.layout.bipartite_layout(G, words_for_graphL.keys()), width = 1,node_size=60,font_size=25)
```
